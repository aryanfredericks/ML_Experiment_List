```python
import numpy as np
```

```python
x= np.array([0.82, 0.31, -0.41])
y= np.array([1])
```

```python
def sigmoid(y):
    return 1/(1+ np.exp(-y))
```

```python
def loss_mse(y_true , y_pred):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
def gradient_descent(x, y , epochs , learning_rate , loss_threshold):
    w = np.zeros(3)
    b= 0
    n = len(y)
    for _ in range(epochs):
        weighted_sum = w[0]*x[0] + w[1]*x[1] + w[2]*x[2] + b
        y_pred = sigmoid(weighted_sum)
        loss = loss_mse(y , y_pred)
        
        if loss < loss_threshold:
            break
        
        dw1 = 1/n * np.sum((y_pred - y) * x[0])
        dw2 = 1/n * np.sum((y_pred - y) * x[1])
        dw3 = 1/n * np.sum((y_pred - y) * x[2])
        db = 1/n * np.sum(y_pred - y)
        w[0] = w[0] - learning_rate * dw1
        w[1] = w[1] - learning_rate * dw2
        w[2] = w[2] - learning_rate * dw3
        b = b - learning_rate * db
        
        print(f"Epoch {_+1} : Loss = {loss} , w1 = {w[0]} , w2 = {w[1]} , w3 = {w[2]} , b = {b}")
    
    return w , b
```

```python
w , b = gradient_descent(x , y , 1000 , 0.1 , 0.01)
```

```python
def predict(x):
    weighted_sum = w[0]*x[0] + w[1]*x[1] + w[2]*x[2] + b
    y_pred = sigmoid(weighted_sum)
    return y_pred

print(predict(np.array([0.5, 0.8, -0.1])))
```

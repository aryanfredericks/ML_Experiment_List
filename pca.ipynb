```python
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler as mms
import matplotlib.pyplot as plt
import pandas as pd
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
min_max_scalar = mms()
X_scaled = min_max_scalar.fit_transform(X)
```

```python
def plot3grapghs(X , title , axis_title):
    plt.figure(figsize=(3,3))
    plt.scatter(X[: , 0] , X[:,1] , c=y)
    plt.title(title)
    plt.xlabel(axis_title + '1')
    plt.ylabel(axis_title + '2')
    plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df = pd.DataFrame(data=  X_pca , columns=['PC1' , 'PC2'])
df.head()
```

```python
plot3grapghs(X_pca , "PCA of Iris Dataset" , "Principal Component")
```

```python
print(f"Explained Variance Ratio is {pca.explained_variance_ratio_}")
```
